```python
from numpy import exp, array, random, dot
import numpy as np
import pandas as pd
from google.colab import drive
from statsmodels.stats.diagnostic import lilliefors
from scipy import stats

class MyNeurona():
    def __init__(self):
      #pesos aleatorios
        self.Pesos_Sinapticos = 2 * random.random([5,1])-1
    def __sigmoide(self, x):
        return 1 / (1 + exp(-x))
    # Derivada de la sigmoide.
    def __derivadaSigmoide(self, x):
        return x * (1 - x)
    # Entrenamiento
    def train(self, entradas_entrenamiento, salidas_entrenamiento, iteraciones):
        for iteration in range(iteraciones):
            # utilizando el conjunto de entrenamiento.
            output = self.think(entradas_entrenamiento)
            # Calculo del error
            error = salidas_entrenamiento - output 
            ajuste = dot(entradas_entrenamiento.T, error * self.__derivadaSigmoide(output))
            # Ajuste de pesos
            self.Pesos_Sinapticos += ajuste

    # Pensamiento
    def think(self, entradas):
        return self.__sigmoide(dot(entradas, self.Pesos_Sinapticos))
```

```python
n=10
#Normalizando datos 
drive.mount('/content/gdrive')
df = pd.read_csv('gdrive/My Drive/Colab Notebooks/MtyVsTigres.csv',header=0,index_col=False)

pivote=0
entrada, salida,salidas, ultimo=[],[],[],[] 
#Calcula salidas
def GEP(pivote):
  if (df['Monterrey'][pivote]>df['Tigres UANL'][pivote]): #Gano
    return 1
  if (df['Monterrey'][pivote]<df['Tigres UANL'][pivote]): #Perdio
    return 0
  else: #Empato 
    return 0.5

#Obteniendo entradas y salidas 
while(pivote!=n):
  dato=[]
  salida.append(GEP(pivote))
  c=pivote+1 #Avanza un partido del pivote
  tgf,tgc=0,0 #total goles favor y contra
  g,p,e=0,0,0 #partidos ganados, perdidos y empatados
  gl,gv,pl,pv,el,ev=0,0,0,0,0,0 #ganados, perdidos y empatados local y visitante
  for i in range(c,c+n): 
    if df['Monterrey'][i]>df['Tigres UANL'][i]:
      g+=1 #Gano
      if df['Rayados'][i]==1:
        gl+=1 #Gano como Local
      else:
        gv+=1 #Gano como visitante
    if df['Monterrey'][i]<df['Tigres UANL'][i]:
      p+=1 #Perdio 
      if df['Rayados'][i]==1:
        pl+=1 #Perdio como Local
      else:
        pv+=1 #Perdio como visitante
    else:
      e+=1 #Empato
      if df['Rayados'][i]==1:
        el+=1 #Empato como local
      else:
        ev+=1 #Empato como visitante
    tgf+=df['Monterrey'][i]
    tgc+=df['Tigres UANL'][i]
  pg=g/n
  pp=p/n
  pgl=gl/(gl+pl+el)
  pgv=gv/(gv+pv+ev)
  gf=tgf/(tgf+tgc)
  dato.append(pg)
  dato.append(pp)
  dato.append(pgl)
  dato.append(pgv)
  dato.append(gf)
  pivote+=1
  entrada.append(dato)
salidas.append(salida)

#Obteniendo la normalizacion del ultimo partido

tgf,tgc=0,0
g,p,e=0,0,0
gl,gv,pl,pv,el,ev=0,0,0,0,0,0
for j in range(n):
  if df['Monterrey'][j]>df['Tigres UANL'][j]:
    g+=1
    if df['Rayados'][j]==1:
      gl+=1
    else:
      gv+=1
  if df['Monterrey'][j]<df['Tigres UANL'][j]:
    p+=1
    if df['Rayados'][j]==1:
      pl+=1
    else:
      pv+=1
  else:
    e+=1
    if df['Rayados'][j]==1:
      el+=1
    else:
      ev+=1
  tgf+=df['Monterrey'][j]
  tgc+=df['Tigres UANL'][j]

ultimo.append(g/n) #Partidos Ganados
ultimo.append(p/n) #Partidos Perdidos
ultimo.append(gl/(gl+pl+el))
ultimo.append(gv/(gv+pv+ev))
ultimo.append(tgf/(tgf+tgc))
print(entrada)
print(salidas)
print(ultimo)
```

```python
pronosticos,iterador=[],0
while(iterador<1000):
  thinks=[]
  for i in range(50):
    redNeuronal = MyNeurona()
    print (f"Pesos aleatorios: {redNeuronal.Pesos_Sinapticos}")
    # Ingresando 4 partidos anteriores
    entradas_entrenamiento = array(entrada)
    salidas_entrenamiento = array(salidas).T
    # Entrenando la red con la muestra de entrenamiento a un determinado ciclo de entrenamiento
    redNeuronal.train(entradas_entrenamiento, salidas_entrenamiento, iterador)
    print (f"Los nuevos pesos sinapticos son: {redNeuronal.Pesos_Sinapticos}")
    # Obteniendo resultado
    print (f"Datos ultimos partidos: {ultimo} ")
    pronostico=redNeuronal.think(array(ultimo))
    print (f"El pronostico de Monterrey vs Tigres UANL es: {pronostico}")
    thinks.append(pronostico[0])
  if iterador==300:
    iterador+=700
  else:
    iterador+=50
  pronosticos.append(thinks)

```
